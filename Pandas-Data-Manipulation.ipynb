#FILTERING - BASED ON CONDITION, GETTING DATA

```python
DF2=df.loc[df["MILES*"]>10,{"START_DATE*","MILES*"}]
DF2
#from column start date
#printing miles series having value greater than 10
```

```python
df1=df.loc[df["MILES*"]>10,["START*","MILES*"]]
df1
#from column start
```

```python
#find out rides from new york
df.loc[df["START*"]=="New York",["MILES*"]]
```

```python
#find out all rides from cary to newyork
#match multiple condt rides from cary and new york-2 cities which means multiple--isin operator
myrides=df[df["START*"].isin(["Cary","New York"])]
myrides.head(n=10)
```

```python
df3=df.loc[(df["MILES*"]>10) & (df["START*"].isin(["Cary","New York"]))]
#class assignmnet : finding all trips of miles greater than 10 and originally starting from cary and new york
```

```python
df3.head(n=20)
```

```python
df3.reset_index()
#index column added and original data is not changed and the( starting is from 0 that what index does)
```

```python
df3.reset_index(inplace=True,drop = True)
df3
#inplace changes original dataframe
#drop removes old index
```
